```python
import numpy as np
```

```python
def scoring(Mat,z,y):
    """
    Function to compute csoring measures E, PE, TFC and FMS.
    Mat: Matching Matrix
    z: vector of feedback in {0,1}
    y: vector of card choices in {1,2,3,4}
    """
    
    # Perseverative Errors 
    
    zz = z
    c = 1
    ee = 1
    start = 1
    
    target_rule_vec = np.array([1,2,3,1,2,3,1,2,3])-1
    i = 0
    
    for t in np.arange(1,len(z)):
        
        if (z[t]==0) & (sum(Mat[t] == y[t])==1):
            perseverative_principle = np.where(Mat[t] == y[t])[0]
            start = t+1
            break
            
        if z[t] == 1:
            if z[t] == z[t-1]:
                c += 1
            if c==10:
                perseverative_principle = 0
                break
                
    c = 1
    
    for t in np.arange(start,len(z)):
        
        if (z[t]==0) & (sum(Mat[t] == y[t])==1):
            if y[t]==Mat[t][perseverative_principle]:
                zz[t] = -1
                
        if z[t]==1:
            if z[t]==z[t-1]:
                c += 1
            if z[t] != z[t-1]:
                c = 1
            if c==10:
                perseverative_principle = target_rule_vec[i]
                i += 1
                c = 1
                
        if ((z[t] == 0) & ((sum(Mat[t] == y[t])==1))) == False:
            ee = 1
        else:
            if (z[t-1] == 0) & (sum(Mat[t-1] == y[t-1])==1):
                if np.where(Mat[t] == y[t])[0] == np.where(Mat[t-1] == y[t-1])[0]:
                    ee += 1
                else:
                    ee = 1
                if ee == 3:
                    perseverative_principle = np.where(Mat[t] == y[t])[0]
                    ee = 1
                        

    for t in np.arange(start,len(z)-1):
        if (zz[t] == 0) & (sum(Mat[t] == y[t])>1):
            if (zz[t-1] == -1) & (zz[t+1] == -1):
                if y[t] == Mat[t][perseverative_principle]:
                    zz[t] = -1
    
    # Trials to First Category
    
    c = 1
    for t in np.arange(1,len(z)):
        if z[t]==1:
            if z[t-1]==1:
                c += 1
            else:
                c = 1
        else:
            c = 1
        if c==10:
            break
    ttfc = t+1
    
    # Failure to Maintain Sest
    
    f = 0
    c = 1
    z = np.append(z,[99,99,99,99,99,99])
    for t in np.arange(1,len(z)-6):
        if (z[t]==1) & (z[t-1]==1):
            c += 1
        else:
            c = 1
        if c==4:
            T = t
            if sum(z[(T+1):(T+6)]==0) > 0:
                f += 1
                c = 1
        if c==10:
            c = 1
                
    return {'E': sum(z==0), 'PE': sum(z==-1), 'TFC': ttfc, 'FMS': f}
        
```
